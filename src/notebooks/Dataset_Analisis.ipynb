## Arreglo de Oracle Text (feat. Gepetto)

```python
import pandas as pd
import ast

# Paso 1: Convertir strings a listas de dicts
def parse_card_faces(card_faces_str):
    try:
        return ast.literal_eval(card_faces_str)
    except (ValueError, SyntaxError):
        return None

df['card_faces'] = df['card_faces'].apply(parse_card_faces)

# Paso 2: Rellenar los oracle_text vac√≠os
def fill_oracle_text(row):
    if pd.isna(row['oracle_text']) and isinstance(row['card_faces'], list):
        texts = [face.get('oracle_text', '').strip() for face in row['card_faces']]
        texts = [t for t in texts if t]  # quitar vac√≠os
        return '\n//\n'.join(texts) if texts else None
    return row['oracle_text']

df['oracle_text'] = df.apply(fill_oracle_text, axis=1)
```

```python
## Ya de paso el mana cost tambi√©n

# Paso 3: Rellenar mana_cost vac√≠o desde card_faces
def fill_mana_cost(row):
    if pd.isna(row['mana_cost']) and isinstance(row['card_faces'], list):
        costs = [face.get('mana_cost', '').strip() for face in row['card_faces']]
        costs = [c for c in costs if c]  # eliminar vac√≠os
        return ' // '.join(costs) if costs else None
    return row['mana_cost']

df['mana_cost'] = df.apply(fill_mana_cost, axis=1)
```
